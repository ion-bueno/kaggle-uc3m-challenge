```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
train_comp = pd.read_csv('raw_data/train_set.csv')
test_comp = pd.read_csv('raw_data/test_set.csv')

X_train = pd.read_csv('preprocessed/X_train_median.csv', index_col=0)
y_train = pd.read_csv('preprocessed/y_train_median.csv', index_col=0)
```

```python
f = plt.figure(figsize=(19, 15))
plt.matshow(X_train.corr(), fignum=f.number)
plt.xticks(range(X_train.select_dtypes(['number']).shape[1]), X_train.select_dtypes(['number']).columns, fontsize=14, rotation=45)
plt.yticks(range(X_train.select_dtypes(['number']).shape[1]), X_train.select_dtypes(['number']).columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16)
```
